```python
%matplotlib inline
#Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Load city and ride data files
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

#Read the City and Ride Data
city_df = pd.read_csv("Resources/city_data.csv")
ride_df = pd.read_csv("Resources/ride_data.csv")
# city_df.head()
# ride_df.head()
```

```python
#Combine the data into a single dataset
ride_city_merge = pd.merge(ride_df, city_df, how="outer", on=["city"])
# ride_city_merge.head(10)
```

```python
#Get unique city types and cities
city_types = ['Urban','Rural','Suburban']
cities = ride_city_merge['city'].unique()
# print(city_types)

#Create a new city dataframe from a copy of the city dataframe.  The new_city dataframe will contain all data related to each city.
new_city_df = city_df 
new_city_df = new_city_df.set_index('city')
# new_city_df.head()
```

```python
#Adding average fare and number of rides for each city to the new_city_df
def retrieve_city_data(x):
    ind_city_df = ride_city_merge.loc[ride_city_merge['city'] == x]
    city_avg_fare = ind_city_df['fare'].mean()
    city_num_rides = ind_city_df['fare'].count()
    city_total_fare = ind_city_df['fare'].sum()
    new_city_df.at[x,'average_fare'] = city_avg_fare
    new_city_df.at[x,'number_of_rides'] = city_num_rides
    new_city_df.at[x,'total_fares_$'] = city_total_fare

for city in cities:
    retrieve_city_data(city)
# new_city_df.head()
```

```python
#Obtain the x and y coordinates for each of the three city types
urban_df = new_city_df.loc[new_city_df['type'] == 'Urban']
suburban_df = new_city_df.loc[new_city_df['type'] == 'Suburban']
rural_df = new_city_df.loc[new_city_df['type'] == 'Rural']
# rural_df.head(100)
```

# Pyber Ride Sharing Analysis
* Urban cities had the highest percentage of number of rides, total fares, and total number of drivers.
* The highest average cost for rides was found in rural towns.
* The number of urban drivers is more than four times greater than the number of suburban and rural drivers combined.

## Bubble Plot of Ride Sharing Data


```python
#Setup chart colors and explode parameters
colors = ["coral", "gold","lightblue"]
explode = (0.1, 0, 0)

#Build the scatter plots for each city types
plt.scatter(urban_df['number_of_rides'],urban_df['average_fare'], marker="o", facecolors="coral", edgecolors="black", s=[urban_df['driver_count']*6], alpha=.85, label="Urban")
plt.scatter(suburban_df['number_of_rides'],suburban_df['average_fare'], marker="o", facecolors="lightblue", edgecolors="black", s=[suburban_df['driver_count']*6],alpha=.85, label="Suburban")
plt.scatter(rural_df['number_of_rides'],rural_df['average_fare'], marker="o", facecolors="gold", edgecolors="black", s=[rural_df['driver_count']*6], alpha=.85, label="Rural")

#Incorporate the other graph properties
plt.hlines(40,0,10,alpha=0.25)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid(axis='both')
# plt.tight_layout()

#Create a legend
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=10)
lgnd.legendHandles[0]._sizes = [40]
lgnd.legendHandles[1]._sizes = [40]
lgnd.legendHandles[2]._sizes = [40]
# plt.show()
           
#Incorporate a text label about the circle size
plt.text(42,35,s="Note:\nCircle size correlates with driver count per city")

#Save Figure
plt.savefig("../Images/pyber-bubble.png", bbox_inches='tight')
```

## Total Fares by City Type

```python
#Calculate Percentages for Fares by City Types
type_pct = []
total_fare = new_city_df['total_fares_$'].sum()
urban_pct = (urban_df['total_fares_$'].sum()/total_fare) * 100
suburban_pct = (suburban_df['total_fares_$'].sum()/total_fare) * 100
rural_pct = (rural_df['total_fares_$'].sum()/total_fare) * 100
type_pct.extend((urban_pct,rural_pct, suburban_pct))

#Build Pie Chart
plt.pie(type_pct, explode=explode, labels=city_types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=300)

#Create axes which are equal
plt.axis("equal")

#Add title
plt.title("% of Total Fares by City Type")

#Save image of chart
plt.savefig("../Images/fares_pct.png")
plt.show()
```

## Total Rides by City Type

```python
#Calculate Percentages of Rides by City Types
type_rides = []
total_rides = new_city_df['number_of_rides'].count()
urban_rides = (urban_df['number_of_rides'].sum()/total_rides) * 100
suburban_rides = (suburban_df['number_of_rides'].sum()/total_rides) * 100
rural_rides = (rural_df['number_of_rides'].sum()/total_rides) * 100
type_rides.extend((urban_rides,rural_rides,suburban_rides))

#Build Pie Chart
plt.pie(type_rides, explode=explode, labels=city_types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=255)

#Create axes which are equal
plt.axis("equal")

#Add title
plt.title("% of Total Rides by City Type")

#Save image of chart
plt.savefig("../Images/rides_pct.png")
plt.show()
```

## Total Drivers by City Type

```python
#Calculate Percentages of Ttoal Drivers by City Types
type_drivers = []
total_drivers = new_city_df['driver_count'].sum()
urban_drivers = (urban_df['driver_count'].sum()/total_drivers) * 100
suburban_drivers = (suburban_df['driver_count'].sum()/total_drivers) * 100
rural_drivers = (rural_df['driver_count'].sum()/total_drivers) * 100
type_drivers.extend((urban_drivers,rural_drivers,suburban_drivers))

#Build Pie Chart
plt.pie(type_drivers, explode=explode, labels=city_types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=225)

#Create axes which are equal
plt.axis("equal")

#Add title
plt.title("% of Total Drivers by City Type")

#Save image of chart
plt.savefig("../Images/drivers_pct.png")
plt.show()
```
